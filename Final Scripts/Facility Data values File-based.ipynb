```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import json
import os
import psycopg2
```

```python
#read from the backup file
def read_file(file_name):
    fileDir = os.path.dirname(os.path.realpath('__file__'))
    filename = os.path.join(fileDir, '../backups/'+file_name)
    with open(filename) as my_file:
        db_conf = json.load(my_file)
    return db_conf
```

```python
#read from the failities table
def get_facilities():
    #data element table analysis
    json_data = read_file('facilities_facility')
    facilities_dataframe = DataFrame(json_data['data']).set_index('index')

    return facilities_dataframe
```

```python
# read from the organization file
def get_organization_units():
    json_data = read_file('dim_dhis_organisationunit')
    org_units_dataframe = DataFrame(json_data['data']).set_index('index')
    
    return org_units_dataframe
```

```python
get_organization_units()
```

```python
def get_facility_org_units(facility_id):
    #get the facility
    all_facilities = get_facilities()
    facility = all_facilities[all_facilities['id'] == facility_id]    
    facility['code'] = facility['code'].astype(str)
    
    #get all org units
    all_orgunits = get_organization_units()
    all_orgunits['code'] = all_orgunits['code'].astype(str)
    facility_orgunits = pd.merge(all_orgunits, facility, on='code')
    
    return facility_orgunits

```

```python
#def get datavalues
def get_data_values():
    #read for the datavalues table
    json_data = read_file('fact_dhis_datavalue')
    data_value_dframe = DataFrame(json_data['data']).set_index('index')
    
    return data_value_dframe
```

```python
get_data_values()
```

```python
def get_facility_data_values(facility_id):
    #get the faciity org units
    facility_orgunits = get_facility_org_units(facility_id)
    
    datavalues_dataframe = get_data_values()
    datavalues_dataframe = datavalues_dataframe.rename(index=str, columns={'sourceid': 'organisationunitid'})
    
    facility_datavalues = pd.merge(facility_orgunits, datavalues_dataframe, on='organisationunitid')
    
    return facility_datavalues
```
