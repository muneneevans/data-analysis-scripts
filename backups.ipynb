```python
def get_connection():
    
    fileDir = os.path.dirname(os.path.realpath('__file__'))
    ilename = os.path.join(fileDir, 'db_connection/db_config.json')
    with open(filename) as my_file:
        db_conf = json.load(my_file)
    conn_str = "host={} dbname={} user={} password={}".format(
                db_conf['host'], db_conf['database'], 
                db_conf['user'], db_conf['passw'])
    
    # return conn_str
    try:
        conn = psycopg2.connect(conn_str)
        return conn
    except Exception:
        return Exception.message
        
get_connection()
```

```python
def get_table():
    conn = get_connection()
    table = pd.DataFrame()
    for chunk in pd.read_sql('select * from dim_dhis_period LIMIT 100000', con=conn, chunksize=100):
        table = table.append(chunk)
    
    table.to_json('backups/dim_dhis_period',orient='table')
    return table

get_table()
```
