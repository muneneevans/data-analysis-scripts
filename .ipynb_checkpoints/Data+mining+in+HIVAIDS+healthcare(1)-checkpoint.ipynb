```python
#transform data types of DOB to date time from string
Patientsdf['DOB'] = pd.to_datetime(Patientsdf['DOB'])
```

```python
#transform data types of Reg at CCC to date time from string
Patientsdf['RegistrationAtCCC'] = pd.to_datetime(Patientsdf['RegistrationAtCCC'])
```

```python
#get age of patients from dataframe
Patientsdf['Age']=(Patientsdf.RegistrationAtCCC - Patientsdf.DOB).astype('<m8[Y]') 
```

```python
#remove decimal places in Age column
Patientsdf['Age']=Patientsdf.Age.astype(int)
```
