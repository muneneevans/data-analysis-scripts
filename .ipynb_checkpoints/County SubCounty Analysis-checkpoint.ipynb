```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import json
import os
import psycopg2
```

```python
#read from the backup file
def read_file(file_name):
    fileDir = os.path.dirname(os.path.realpath('__file__'))
    filename = os.path.join(fileDir, 'backups/'+file_name)
    with open(filename) as my_file:
        db_conf = json.load(my_file)
    return db_conf
```

```python
#facilities table analysis
json_data = read_file('facilities_facility')
facilities_dataframe = DataFrame(json_data['data']).set_index('index')

facilities_dataframe.dtypes
# facilities_dataframe[['dataelementname','description']]
```

```python
def get_county_constituencies(county_id):
    json_data = read_file('common_constituency')
    constituencies_dataframe = DataFrame(json_data['data']).set_index('index')
    county = counties_dataframe[counties_dataframe['id']==county_id]
    county = county.rename(index=str, columns={"id": "county_id", 'name': 'county_name'})
    county_constituencies = pd.merge(constituencies_dataframe,county, on='county_id')
```
