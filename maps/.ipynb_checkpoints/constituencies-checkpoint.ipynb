```python
#import needed libraries
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import json
import psycopg2
import os
 
        
```

```python
def get_kenya_constituenceis():
    with open('maps/constituencies.json') as my_file:
        map_json = json.load(my_file)
    
    for constituency in map_json['objects']['constituencies']['geometries']:
        file_name = "maps/{}.json".format(constituency['properties']['CONSTITUEN'])
        print(constituency)
        file = open(file_name,'w') 
        file.write(constituency) 
```

```python
get_kenya_constituenceis()
```
