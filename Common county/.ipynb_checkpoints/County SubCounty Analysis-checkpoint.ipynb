```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import json
import os
import psycopg2
```

```python
#read from the backup file
def read_file(file_name):
    fileDir = os.path.dirname(os.path.realpath('__file__'))
    filename = os.path.join(fileDir, '../backups/'+file_name)
    with open(filename) as my_file:
        db_conf = json.load(my_file)
    return db_conf
```

```python
#get all constituencies in a county by id
def get_county_constituencies(county_id):
    json_data = read_file('common_constituency')
    constituencies_dataframe = DataFrame(json_data['data']).set_index('index')
    county = counties_dataframe[counties_dataframe['id']==county_id]
    county = county.rename(index=str, columns={"id": "county_id", 'name': 'county_name'})
    county_constituencies = pd.merge(constituencies_dataframe,county, on='county_id')
    county_constituencies = county_constituencies[['name','id']]
    
    return county_constituencies
```

```python
def get_constituency_wards_ids_json(constituency_id):
    '''return ids for wards in a constituency'''
    all_wards = read_file('common_ward')
    
    constituency = read_file('common_constituency')
#     constituency = constituency.rename(index=str, columns={'id':'constituency_id', 'name': "constituency_name"})

    constituency_wards = pd.merge(all_wards,constituency, on='constituency_id')
    constituency_wards = constituency_wards[['name','id','constituency_id']]

    return constituency_wards.to_json(orient='records')

get_constituency_wards_ids_json('f75a7118-107f-4c12-9236-dc460d3e33df')
```
